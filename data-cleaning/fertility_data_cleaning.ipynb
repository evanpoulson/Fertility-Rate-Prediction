```python
import pandas as pd
import numpy as np
```

```python
df_fetility = pd.read_csv('API_SP.DYN.TFRT.IN_DS2_EN_csv_v2_129476.csv')
df_urban = pd.read_csv('API_SP.URB.TOTL.IN.ZS_DS2_en_csv_v2_129596.csv')
df_GDP = pd.read_csv('API_NY.GDP.PCAP.CD_DS2_en_csv_v2_248635.csv')
```

```python
drop_countries = regional_codes = [
    'AFE', 'AFW', 'ARB', 'CEB', 'CSS', 'EAP', 'EAR', 'EAS', 'ECA', 'ECS', 
    'EMU', 'EUU', 'FCS', 'HIC', 'HPC', 'IBD', 'IBT', 'IDA', 'IDB', 'IDX',
    'INX', 'LAC', 'LCN', 'LDC', 'LIC', 'LMC', 'LMY', 'LTE', 'MEA', 'MIC',
    'MNA', 'NAC', 'OED', 'OSS', 'PRE', 'PSS', 'PST', 'SAS', 'SSA', 'SSF',
    'SST', 'TEA', 'TEC', 'TLA', 'TMN', 'TSA', 'TSS', 'UMC', 'WLD'
] 

df_fertility_edited = df_fetility[~df_fetility['Country Code'].isin(drop_countries)]
display(df_fertility_edited[df_fertility_edited['Country Code'].isin(drop_countries)])
```

```python
df_urban = df_urban.drop(columns=['Indicator Name', 'Indicator Code','Unnamed: 69' ])
df_GDP = df_GDP.drop(columns=['Indicator Name', 'Indicator Code','Unnamed: 69' ])
df_fertility_edited = df_fertility_edited.drop(columns=['Indicator Name', 'Indicator Code','Unnamed: 69','2024' ])
```

```python
df_urban_edited = df_urban.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', 
value_name='Urban population over total population')

df_fertility_edited = df_fertility_edited.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', 
value_name='Fertirity rate')

df_GDP_edited = df_GDP.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', 
value_name='Capitial GDP in USD')
```

```python
foreign_key = ['Country Name', 'Year', 'Country Code']
df_merged = df_fertility_edited.merge(df_urban_edited, on = foreign_key , how='left') \
                                    .merge(df_GDP_edited, on = foreign_key, how='left')
```

```python
df_merged.to_csv('Fertirity_GDP_Urban_Country_Year_1.csv', index = False)
```
